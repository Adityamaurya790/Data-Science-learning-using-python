### First import all libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## some plots that allow us to visualize the distribution of the data set.Types of plotes are:
* histplot/distplot
* jointplot
* pairplot
* rugplot

### Load dataset using seaborn

```python
df = sns.load_dataset('tips')
```

```python
df
```

```python
df['size'].unique()
```

### Histogram plot

```python
plt1 = plt.subplot(2,2,1)
plt1 = sns.histplot(df['total_bill'],color='red',bins=50,kde=True)

plt2 = plt.subplot(2,2,4)
plt2 = sns.histplot(df['tip'],kde=True)

# plt3 = plt.subplot(2,2,3)
# plt3 = sns.histplot(df['day'])
```

### Joint Plot

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind ='scatter', hue='sex' )
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df, kind ='hex')
```

```python
sns.pairplot(df,hue = 'sex')
```

### Pair Plot

```python
sns.pairplot(df,hue = 'size',palette='rainbow')
```

### Rug plot

```python
sns.rugplot(df)
```
