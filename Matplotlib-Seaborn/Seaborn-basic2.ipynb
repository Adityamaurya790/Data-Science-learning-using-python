```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# categorical data ploting 
#### categorial means the column has a category like (make,female)(true,false)(yes,no) etc

* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
df = sns.load_dataset('tips')
df
```

### Countplot

```python
sns.countplot(x = df['sex'],hue=df['smoker'],palette='rainbow')
```

### barplot

```python
sns.barplot(x= df['sex'],y = df['total_bill'])
```

### comparision of total tip given by male and female

```python
import numpy as np
sns.barplot(x=df['sex'],y=df['tip'],estimator=np.sum)
```

```python

sns.barplot(x=df['sex'],y=df['total_bill'],estimator=np.sum)
```

## Boxplot

```python
sns.boxplot(x='tip',y='day',data=df)
```

### Violinplot

```python
sns.violinplot(y='tip',x='day',data=df)
```

### stripplot and swarmplot

```python
sns.stripplot(x='tip',y='day',data=df)
```

```python
sns.swarmplot(x='day',y='tip',data=df,palette='rainbow')
```

### creating Violinplot and swarmplot at same canvas

```python
sns.violinplot(y='tip',x='day',data=df)
sns.swarmplot(x='day',y='tip',data=df)
```
