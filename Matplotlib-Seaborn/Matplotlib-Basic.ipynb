```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1,5,11)
x
```

```python
y = x**2
y
```

```python
plt.plot(x,y)
plt.title('Basic') #title function is use to name the graph 
plt.xlabel('X Ki value')
plt.ylabel('y ki value')
```

```python
# Subplots
val1= plt.subplot(2,2,1)
val1.plot(x,y)

val2= plt.subplot(2,2,4)
val2.plot(x,x**9)
```

#### object oriented way

```python
fig = plt.figure(dpi=100,facecolor='yellow',frameon=True)
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)

axis2=fig.add_axes([0.2,0.4,0.3,0.3])
axis2.plot(y,x)
```
