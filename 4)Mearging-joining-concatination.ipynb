```python
import numpy as np
import pandas as pd
```

## üëâ Merging two dataframes 

```python
# employees and salaries data are benn taken from AkarshVyas github

# employees is -> 1st dataframe
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# salaries is -> 2nd dataframe
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
employees
```

```python
salaries
```

```python
 pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
# -> if we do how=outer it will merge all the data from both dataframes and provide NAN where data will be missing
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
# letf -> means all data will be priented from employees and NAN to pta hi hai
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
# Right -> means all data will be priented from employees
pd.merge(employees,salaries,on='employee_id',how='right')
```

## üëâConcatination in dataframes

```python
# creating two data frame and the concatinatiog them 
# this data are taken from AkarshVyas github

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2'],
    'B': ['B0', 'B1', 'B2'],
    'C': ['C0', 'C1', 'C2']
})

df2 = pd.DataFrame({
    'A': ['A3', 'A4', 'A5'],
    'B': ['B3', 'B4', 'B5'],
    'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) # this is done on column based
```

```python
pd.concat([df1,df2],axis =1) # this is done on row based
```

```python
pd.concat([employees,salaries],axis =1) # like here it has added employees and salaries in row wise
```

```python
pd.concat([employees,salaries]) # like here it has added employees and salaries in column wise
```

## üëâ Joining in pandas
