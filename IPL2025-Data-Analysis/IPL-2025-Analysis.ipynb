### First we have to load libraries and data set to our notebook

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import warnings 

# warnings.filterwarnings("ignore")

ipl = pd.read_csv('IPL25.csv')
ipl
````

### Droping Match_id because index is there so no need for that

````python
ipl.drop('match_id',inplace=True,axis = 1)
ipl
````

### filling all null value

````python
ipl["margin"].fillna(0, inplace=True)
ipl['margin_type'].fillna(0,inplace = True)
````

````python
ipl.isnull().sum()
````

### Q1) Which team won the most matches?

````python
ipl["match_winner"].value_counts()
````

````python
match_wins = ipl['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x = match_wins.values,palette='viridis')
plt.title("Most match win by team ")
````

### How do teams win? (Runs vs Wickets)

````python
sns.countplot(x = ipl['won_by'])
plt.title("Won by")
````
